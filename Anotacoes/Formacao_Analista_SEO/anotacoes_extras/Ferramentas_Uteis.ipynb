# Ferramentas Uteis de SEO

- Para simular o preview da SERP com nosso title e meta description: [SERP Preview Tool](https://blog.spotibo.com/serp-preview-tool/);

- Para simular Rich Snipets baseados nos dados estruturados de uma página: [Teste de Pesquisa Aprimorada](https://search.google.com/test/rich-results);

- Site para aprender a criar e entender melhor `sitemaps` é: [sitemaps.org](https://sitemaps.org/);

- Site para aprender a criar e entender melhor sobre dados estruturados: [schema.org](https://schema.org/docs/gs.html);

- Ferramenta para criação de esquemas e design da estrutura/arquitetura de um site: [Figma](https://www.figma.com/) (Pode ser substituída por uma planilha no Excel, o que é muito menos apresentável visualmente falando);

- Contador de palavras, linhas e palavras: [Contador de Caracteres](https://www.invertexto.com/contador-caracteres);

- Validador de HTML e CSS: [Validator](https://validator.w3.org/);

- Ferramenta para testar protocolo HTTP: [KeyCDN](https://tools.keycdn.com/http2-test);

- Ferramente para minificação de documentos CSS ou JS: [Minifier](https://www.minifier.org/);

- Ferramenta que ajuda a montar design systems: [PatternLab](https://patternlab.io/);

- Ferramenta para compressão de imagens: [ILoveIMG](https://www.iloveimg.com/pt);

- Plugins para compressão, conversão e disponibilização de imagens em WordPress: [Optimole](https://wordpress.org/plugins/optimole-wp/) ou [Smush](https://br.wordpress.org/plugins/wp-smushit/)

- Ferramentas de performance:
    - [Merkle](https://technicalseo.com/tools/):
        - Destacando as ferramentas de Render e Pre-Render.
    - [Smartlook](https://www.smartlook.com/);
        - Ferramenta para análise de comportamento de usuários dentro do site.
    - [SEO Pro Extension](https://chrome.google.com/webstore/detail/seo-pro-extension/lajfdngpocmchpihjecojjllfjeehgnl):
        - É a extensão do Chrome que é completinha, que traz o `title`, `cannonical`, e outras métricas do Core Web Vitals.
    - [Bing webmaster](https://www.bing.com/webmasters/about);
    - [Semrush](https://pt.semrush.com/);
    - [WebPageTest](https://www.webpagetest.org/);
    - [Google PageSpeed](https://pagespeed.web.dev/);
    - [Dareboost](https://www.dareboost.com/en);
        - Essa ferramenta foi comprada e provavelmente sairá do mercado;
        - Foi considerada por muitos a melhor ferramenta para otimizar arquivos JS e CSS "no detalhe".
    - [GTmetrix](https://gtmetrix.com/).

- Monitoramento de UPTIME:
    - [UptimeRobot](https://uptimerobot.com/);
    - [pingdom](https://www.pingdom.com/);
- CDN:
    - [Cloudflare](https://www.cloudflare.com/pt-br/);
    - [gocache](https://www.gocache.com.br/);
- Indicação de server de confiança:
    - [Hostinger](https://www.hostinger.com/) (Cloud ou VPS);
    - [WP Engine](https://wpengine.com/) (Hospedagem WordPress);
    - [AWS](https://aws.amazon.com/pt/) (Cloud + CDN);
    - [DigitalOcean](https://www.digitalocean.com/) (Cloud VPN);
